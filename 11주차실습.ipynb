```python
# 텍스트에서 특수문자, 숫자 등을 제거하고 영문을 소문자로 변환하여 정규화합니다.
import re

def normalize_text(text):
  text = re.sub(r'[^ㄱ-ㅎ|가-힣| ]', '', text)
  text = text.lower()
  return text
```

```python
# 예시 문장
sample_text = "안녕하세요! 이 영화 정말 쵝오!! 2024년 최고의 영화입니다. This is a GREAT movie. :) ㅋㅋㅋ"

print("### 텍스트 정규화 전후 비교 ###")
print(f"원본 텍스트: {sample_text}")
normalized_sample = normalize_text(sample_text)
print(f"정규화 후 텍스트: {normalized_sample}")
```

```python
# 학습 목표: KoNLPy의 Okt 형태소 분석기를 사용하여 명사, 동사, 형용사만 추출하여 텍스트를 토큰화합니다.
from konlpy.tag import Okt

okt = Okt() # Okt 형태소 분석기 초기화

def tokenize_text(text):
  pos_results = okt.pos(text, norm=True, stem=True)
  target_tags = ['Noun', 'Verb', 'Adjective'] # 원하는 품사만 필터링
  words = [word for word, tag in pos_results if tag in target_tags]
  return words
```

```python
# 예시 문장 (정규화된 텍스트 사용 -> normalized_sample)

print("### 형태소 분석 및 토큰화 전후 비교 ###")
print(f"정규화된 텍스트: {normalized_sample}")
tokenized_sample = tokenize_text(normalized_sample)
print(f"토큰화 후 리스트: {tokenized_sample}")
```

```python
# 정의된 불용어 리스트를 사용하여 토큰에서 불용어를 제거합니다.

# 불용어 리스트 정의
stopwords = ['이', '그', '저', '것', '수', '때', '등', '어', '하다', '되다', '이다',
             '있다', '없다', '않다', '같다', '말', '안', '좀', '정말', '진짜', '너무',
             '가장', '더', '또', '잘', '못', '내', '네', '알', '아', '다', '만들다', '같다']
# TODO: 불용어 리스트에 자주 사용되지만 의미가 없는 단어들을 5개 이상 추가해 보세요.
stopwords.extend(['영화', '드라마', '생각', '사람', '보고', '하나', '속', '점'])

def remove_stopwords(tokens):
  filtered_tokens = [
      word for word in tokens
      if word not in stopwords and len(word) > 1
  ]
  return filtered_tokens
```

```python
print("### 불용어 제거 전후 비교 ###")
print(f"토큰화 후 리스트: {tokenized_sample}")
filtered_sample = remove_stopwords(tokenized_sample)
print(f"불용어 제거 후 리스트: {filtered_sample}")
```

```python
# 텍스트 정규화, 형태소 분석 및 토큰화, 불용어 제거를 하나의 통합 함수로 만듭니다.
def preprocess_text(text):
  if not isinstance(text, str): # 텍스트가 아닌 경우 빈 리스트 반환
    return []
  normalized_text = normalize_text(text) # 텍스트 정규화
  tokenized_text = tokenize_text(normalized_text) # 형태소 분석 및 토큰화
  final_tokens = remove_stopwords(tokenized_text) # 불용어 제거
  return final_tokens
```

```python
# 전체 데이터에 적용하여 전처리된 결과를 생성하고 저장합니다.

# 학습 데이터에 전처리 함수 적용 (tqdm으로 진행률 표시)
print("학습 데이터(train_df) 전처리 중...")
train_df["processed_document"] = train_df["document"].progress_apply(preprocess_text)

# 테스트 데이터에 전처리 함수 적용 (tqdm으로 진행률 표시)
print("테스트 데이터(test_df) 전처리 중...")
test_df["processed_document"] = test_df["document"].progress_apply(preprocess_text)

print("전처리 완료된 학습 데이터 확인(상위5개)")
display(train_df.head())

# 전처리 완료된 데이터 저장(CSV)
train_df.to_csv('ratings_train_processed.csv', index=False, encoding="utf-8")
test_df.to_csv('ratings_test_processed.csv', index=False, encoding="utf-8")
```

# 2차시: BoW, DTM, TF-IDF

```python
train_df = pd.read_csv('ratings_train_processed.csv')
test_df = pd.read_csv('ratings_test_processed.csv')

# 'processed_document' 컬럼은 문자열 리스트 형태로 저장되었으므로, 다시 리스트로 변환합니다.
# NaN 값은 빈 리스트로 처리하여 오류를 방지합니다.
train_df['processed_document'] = train_df['processed_document'].apply(
    lambda x: eval(x) if pd.notna(x) else [])
test_df['processed_document'] = test_df['processed_document'].apply(
    lambda x: eval(x) if pd.notna(x) else [])

print(f"학습 데이터(train_df) 정보({train_df.shape})")
display(train_df.head())

print(f"\n테스트 데이터(test_df) 정보({test_df.shape})")
display(test_df.head())
```

```python
# CountVectorizer를 사용하여 BoW(Bag of Words) 방식으로 텍스트를 벡터화
import pickle
from scipy.sparse import save_npz, load_npz
from sklearn.feature_extraction.text import CountVectorizer

# processed_document 컬럼의 리스트를 공백으로 연결된 문자열로 변환
train_docs_str = [' '.join(doc) for doc in train_df["processed_document"]]
test_docs_str = [' '.join(doc) for doc in test_df["processed_document"]]

bow_vectorizer = CountVectorizer(min_df=5, max_features=10000)
bow_matrix = bow_vectorizer.fit_transform(train_docs_str)

print("BoW 벡터화 결과")
print("단어 사전(vocabulary) 크기:", len(bow_vectorizer.vocabulary_))
print("BoW 행렬 구조(shape):", bow_matrix.shape)
print("BoW 행렬 타입:", type(bow_matrix)) # 희소 행렬(sparse matrix) 확인

# 행렬 저장
save_npz('bow_train_matrix.npz', bow_matrix)
print("\nbow_train_matrix.npz 저장 완료")

# 벡터라이저 저장 (테스트 데이터 변환 시 필요)
with open('bow_vectorizer.pkl', 'wb') as f:
    pickle.dump(bow_vectorizer, f)
print("bow_vectorizer.pkl 저장 완료")
```

```python
# 가장 많이 등장한 상위 20개 단어 확인
import numpy as np
word_counts = np.array(bow_matrix.sum(axis=0)).flatten()
feature_names = bow_vectorizer.get_feature_names_out()

# 단어와 해당 빈도를 매핑하고 빈도 기준으로 정렬
word_freq = sorted(list(zip(feature_names, word_counts)), key=lambda x: x[1], reverse=True)

print("상위 20개 단어와 빈도")
for word, count in word_freq[:20]:
    print(f"\t{word}: {count}")
```

```python
# 샘플 문서로 DTM(Document-Term Matrix)을 직접 구성하고 그 구조를 시각화

# 샘플 문서 5개 선택
sample_docs = ["영화 재미있다 좋다", "정말 재미있다 최고다", "별로 재미없다", "너무 좋다 추천하다", "정말 별로"]

sample_vectorizer = CountVectorizer()
sample_dtm = sample_vectorizer.fit_transform(sample_docs)

# DTM을 DataFrame으로 변환하여 시각화
sample_dtm_df = pd.DataFrame(sample_dtm.toarray(),
                             columns=sample_vectorizer.get_feature_names_out(),
                             index=[
                                 f'문서{i+1}' for i in range(len(sample_docs))
                                 ]
                             )
print("샘플 DTM:")
display(sample_dtm_df)
```

```python
print("\n전체 데이터 DTM 크기 및 희소 행렬 특성 확인")
print("전체 DTM 구조(shape):", bow_matrix.shape)
```

```python
# DTM의 희소성(sparsity) 계산

total_elements = bow_matrix.shape[0] * bow_matrix.shape[1]
zero_elements = total_elements - bow_matrix.nnz # nzz = number of non-zero elements
sparsity = (zero_elements / total_elements) * 100
print(f"DTM의 희소성: {sparsity:.2f}")
```

```python
# TfidfVectorizer를 사용하여 TF-IDF(Term Frequency-Inverse Document Frequency)를 계산
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(min_df=5, max_features=10000, sublinear_tf=True)
tfidf_matrix = tfidf_vectorizer.fit_transform(train_docs_str)

print("TF-IDF 행렬 구조(shape):", tfidf_matrix.shape)
print("TF-IDF 행렬 타입:", type(tfidf_matrix))

# 저장
save_npz('tfidf_train_matrix.npz', tfidf_matrix)
print("\ntfidf_train_matrix.npz 저장 완료")

# TF-IDF 벡터라이저 객체 저장 (나중에 새로운 텍스트를 변환할 때 사용)
with open('tfidf_vectorizer.pkl', 'wb') as f:
    pickle.dump(tfidf_vectorizer, f)
```

```python
# 샘플 문서 (DTM 구성 시 사용한 것과 동일)
sample_docs_manual = [
    "영화 재미있다 좋다",
    "정말 재미있다 최고다",
    "별로 재미없다",
    "너무 좋다 추천하다",
    "정말 별로"
]

# 단어 사전 (샘플 문서들로부터)
feature_names_manual = sorted(list(set(' '.join(sample_docs_manual).split())))

# 1. TF (Term Frequency) 계산 함수
def calculate_tf(doc, term):
    # TF = (단어 t가 문서 d에 나타나는 횟수) / (문서 d의 총 단어 수)
    doc_tokens = doc.split()
    if len(doc_tokens) == 0: return 0
    return doc_tokens.count(term) / len(doc_tokens)

# 2. IDF (Inverse Document Frequency) 계산 함수
def calculate_idf(docs, term):
    # N: 전체 문서 수, df(t): 단어 t를 포함하는 문서의 수
    # IDF = log(N / (df(t) + 1))  (분모에 1을 더하여 0으로 나누는 것을 방지)
    N = len(docs)
    df_t = sum(1 for doc in docs if term in doc.split())
    return np.log(N / (df_t + 1))

# TF-IDF 값 저장할 딕셔너리
tfidf_manual_results = []

# 각 문서에 대해 TF-IDF 계산
for i, doc in enumerate(sample_docs_manual):
    doc_tfidf = {}
    for term in feature_names_manual:
        tf = calculate_tf(doc, term)
        idf = calculate_idf(sample_docs_manual, term)
        tfidf = tf * idf
        doc_tfidf[term] = tfidf
    tfidf_manual_results.append(doc_tfidf)

# 계산 결과를 DataFrame으로 출력
tfidf_manual_df = pd.DataFrame(tfidf_manual_results, index=[f'문서{i+1}' for i in range(len(sample_docs_manual))])
print("직접 계산한 샘플 TF-IDF:")
display(tfidf_manual_df)

# TfidfVectorizer로 동일 샘플 문서 변환하여 결과 비교
sample_tfidf_matrix_sklearn = tfidf_vectorizer.transform(sample_docs_manual)
sample_tfidf_df_sklearn = pd.DataFrame(sample_tfidf_matrix_sklearn.toarray(), columns=tfidf_vectorizer.get_feature_names_out(), index=[f'문서{i+1}' for i in range(len(sample_docs_manual))])

print("\nSklearn TfidfVectorizer로 계산한 샘플 TF-IDF (일부 단어):")
# 비교를 위해 직접 계산한 단어들 위주로 출력
display(sample_tfidf_df_sklearn[[col for col in feature_names_manual if col in sample_tfidf_df_sklearn.columns]])
```

```python
# BoW(Bag of Words) 기반으로 나이브 베이즈 감성 분류 모델을 학습.

print("BoW 기반 나이브 베이즈 모델 학습")

# bow_train_matrix는 이미 로드했음
y = train_df_processed["label"]

X_train_bow, X_test_bow, y_train_bow, y_test_bow = train_test_split(
    bow_train_matrix, y, test_size=0.2, random_state=42, stratify=y
)

print(f"BoW 학습 데이터 shape: {X_train_bow.shape}")
print(f"BoW 테스트 데이터 shape: {X_test_bow.shape}")
```

```python
# DTM(Document-Term Matrix) 기반으로 나이브 베이즈 감성 분류 모델을 학습
print("DTM 기반 나이브 베이즈 모델 학습")

# dtm_train_matrix은 이미 로드하였음
y = train_df_processed["label"]

X_train_dtm, X_test_dtm, y_train_dtm, y_test_dtm = train_test_split(
    dtm_train_matrix, y, test_size=0.2, random_state=42, stratify=y
)

print(f"DTM 학습 데이터 shape: {X_train_dtm.shape}")
print(f"DTM 테스트 데이터 shape: {X_test_dtm.shape}")
```

```python
# Multinomial Naive Bayes 모델 초기화 및 학습
mnb_dtm_model = MultinomialNB()
mnb_dtm_model.fit(X_train_dtm, y_train_dtm)
y_pred_dtm = mnb_dtm_model.predict(X_test_dtm) # 예측

accuracy_dtm = accuracy_score(y_test_dtm, y_pred_dtm)
print(f"DTM 기반 나이브 베이즈 모델의 정확도: {accuracy_dtm:.4f}")
```

```python
# TF-IDF 기반으로 나이브 베이즈 감성 분류 모델을 학습

print("TF-IDF 기반 나이브 베이즈 모델 학습")

# tfidf_train_matrix는 이미 로드하였음.
y = train_df_processed["label"]

X_train_tfidf, X_test_tfidf, y_train_tfidf, y_test_tfidf = train_test_split(
    tfidf_train_matrix, y, test_size=0.2, random_state=42, stratify=y
  )

print(f"TF-IDF 학습 데이터 shape: {X_train_tfidf.shape}")
print(f"TF-IDF 테스트 데이터 shape: {X_test_tfidf.shape}")
```

```python
# Multinomial Naive Bayes 모델 초기화 및 학습
mnb_tfidf_model = MultinomialNB()
mnb_tfidf_model.fit(X_train_tfidf, y_train_tfidf)
y_pred_tfidf = mnb_tfidf_model.predict(X_test_tfidf)

# 정확도 평가
accuracy_tfidf = accuracy_score(y_test_tfidf, y_pred_tfidf)
print(f"TF-IDF 기반 나이브 베이즈 모델 정확도: {accuracy_tfidf:.4f}")
```

```python
# BoW, DTM, TF-IDF 수치화 방식에 따른 나이브 베이즈 모델의 성능을 비교
model_accuracies = {
    "BoW": accuracy_bow,
    "DTM": accuracy_dtm,
    "TF-IDF": accuracy_tfidf
}

accuracy_df = pd.DataFrame(list(model_accuracies.items()), columns=["수치화 방식", "정확도"])
print("모델별 정확도")
display(accuracy_df)
```

```python
# 전처리 여부에 따른 나이브 베이즈 모델의 성능을 비교.

print("### 전처리 방식에 따른 나이브 베이즈 모델 성능 비교 ###")

# --- 전처리 없는 데이터에 대한 TF-IDF 벡터화 및 모델 학습 ---
print("\n--- 전처리 없는 데이터에 TF-IDF 적용 및 모델 학습 중 ---")

# 원본 'document' 컬럼 사용 (NaN 값 처리)
raw_train_docs = train_df_processed['document'].fillna('').tolist()
raw_test_docs = test_df_processed['document'].fillna('').tolist()

# 전처리 없는 TF-IDF 벡터라이저 초기화 (하이퍼파라미터는 이전과 유사하게 설정)
# min_df, max_features, sublinear_tf 등은 데이터에 따라 조정 필요
raw_tfidf_vectorizer = TfidfVectorizer(min_df=5, max_features=10000, sublinear_tf=True)

# 원본 학습 데이터에 TF-IDF 모델 학습 및 변환
raw_tfidf_train_matrix = raw_tfidf_vectorizer.fit_transform(raw_train_docs)

# 원본 테스트 데이터 변환
raw_tfidf_test_matrix = raw_tfidf_vectorizer.transform(raw_test_docs)

# 학습 데이터의 레이블
y = train_df_processed['label']

# 학습/테스트 데이터 분리 (원본 TF-IDF 행렬과 레이블)
# stratify=y를 사용하여 레이블 분포를 유지하면서 분리합니다.
X_train_raw_tfidf, X_test_raw_tfidf, y_train_raw_tfidf, y_test_raw_tfidf = train_test_split(raw_tfidf_train_matrix, y, test_size=0.2, random_state=42, stratify=y)

# Multinomial Naive Bayes 모델 초기화 및 학습
mnb_raw_tfidf_model = MultinomialNB()
mnb_raw_tfidf_model.fit(X_train_raw_tfidf, y_train_raw_tfidf)

# 예측
y_pred_raw_tfidf = mnb_raw_tfidf_model.predict(X_test_raw_tfidf)

# 정확도 평가
accuracy_raw_tfidf = accuracy_score(y_test_raw_tfidf, y_pred_raw_tfidf)
print(f"전처리 없는 TF-IDF 기반 나이브 베이즈 모델 정확도: {accuracy_raw_tfidf:.4f}")

# --- 전처리 유무에 따른 정확도 비교 ---
model_accuracies_preprocessing = {
    '전처리 포함 TF-IDF': accuracy_tfidf,
    '전처리 없음 TF-IDF': accuracy_raw_tfidf
}

accuracy_preprocessing_df = pd.DataFrame(list(model_accuracies_preprocessing.items()), columns=['전처리 방식', '정확도'])
print("\n--- 전처리 유무 모델별 정확도 비교 ---")
display(accuracy_preprocessing_df)

# 막대 그래프로 시각화
plt.figure(figsize=(8, 5))
sns.barplot(x='전처리 방식', y='정확도', data=accuracy_preprocessing_df, palette='cubehelix')
plt.title('전처리 유무에 따른 나이브 베이즈 모델 정확도 비교')
plt.ylim(0.75, 0.9) # 정확도 비교를 위해 y축 범위 설정
plt.xlabel('텍스트 전처리 방식')
plt.ylabel('정확도')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 새로운 리뷰를 TF-IDF 기반 모델로 입력하여 긍정/부정 감성을 예측

new_review = "이 영화 진짜 시간 아깝네요.. 보는 내내 지루하고 짜증만 났습니다."
print(f"입력 리뷰: {new_review}")

# 1. 리뷰 전처리
processed_new_review_tokens = preprocess_text(new_review)
processed_new_review_str = ' '.join(processed_new_review_tokens)
print(f"전처리된 리뷰: {processed_new_review_str}")

# 2. TF-IDF 벡터화
new_review_tfidf = tfidf_vectorizer.transform([processed_new_review_str])

# 3. 모델 예측
prediction = mnb_tfidf_model.predict(new_review_tfidf)
prediction_proba = mnb_tfidf_model.predict_proba(new_review_tfidf)
```

```python
sentiment_map = {0: '부정 (Negative)', 1: '긍정 (Positive)'}
predicted_sentiment = sentiment_map[prediction[0]]
positive_proba = prediction_proba[0][1] # 긍정 확률
negative_proba = prediction_proba[0][0] # 부정 확률

print(f"\n예측 감성: {predicted_sentiment}")
print(f"긍정 확률: {positive_proba:.4f}")
print(f"부정 확률: {negative_proba:.4f}")
```
